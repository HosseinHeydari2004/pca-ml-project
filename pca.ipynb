```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as  sns
```

```python
data = pd.read_csv("Country-data.csv")
```

```python
df = pd.DataFrame(data=data)
```

```python
for col in df.columns:
    print(f"unique values {col}:")
    print(f"{df[col].unique()}")
    print("_"*80)
```

```python
def hist(x, title, x_title):
    sns.set_style(style="whitegrid")
    plt.figure(figsize=(20, 10), dpi=120)
    sns.histplot(data=df, x=x, bins="auto", alpha=1)
    plt.title(label=title, fontsize=20)
    plt.xlabel(xlabel=x_title, fontsize=18)
    plt.ylabel(ylabel="count", fontsize=18)
    plt.tick_params(axis="both", labelsize=17)
    plt.show()
```

```python
hist(x=df["child_mort"], title="histogram of child_mort", x_title="child_mort")
```

â˜‘ï¸ The histogram of child_mort shows that:
1. The graph is skewed to the right
2. Most countries are between 0 and 40 (almost most countries have good to very good health and well-being)
3. There is an outlier at 200

```python
hist(x=df["exports"], title="histogram exports", x_title="exports")
```

â˜‘ï¸ The exports histogram shows that:
1. The graph is skewed to the right
2. Most countries have exports less than 25% (domestic economy)
3. Some countries have between 25% and 75% (moderate to export economy)
4. Some countries have above 75% (export-dependent economies)
5. How many countries have above 100, 150, 175, and 200% (export-dependent to highly foreign trade-dependent economies)

```python
hist(x=df["health"], title="histogram of health", x_title="health")
```

âœ… The exports histogram shows that:
1. The graph has a normal distribution with a slight skew to the right
