```python
def hist(x, title, x_title):
    sns.set_style(style="whitegrid")
    plt.figure(figsize=(20, 10), dpi=120)
    sns.histplot(data=df, x=x, bins="auto", alpha=1)
    plt.title(label=title, fontsize=20)
    plt.xlabel(xlabel=x_title, fontsize=18)
    plt.ylabel(ylabel="count", fontsize=18)
    plt.tick_params(axis="both", labelsize=17)
    plt.show()
```

```python
hist(x=df["child_mort"], title="histogram of child_mort", x_title="child_mort")
```

2. Most countries are between 0 and 40 (almost most countries have good to very good health and well-being)

```python
sns.set_style(style="whitegrid")
plt.figure(figsize=(20, 10), dpi=120)
sns.scatterplot(data=df, x="gdpp", y="life_expec", s=200)
plt.title(
    label="The relationship between countries' annual income and per capita GDP",
    fontsize=20,
)
plt.xlabel(xlabel="gdpp", fontsize=18)
plt.ylabel(ylabel="life_expec", fontsize=18)
plt.tick_params(axis="both", labelsize=17)
plt.show()
```

âœ…General pattern:
1. Clear positive relationship:
    - As GDP per capita increases, life expectancy also increases
2. Diminishing returns effect:<br>
Up to a certain point, increasing GDP increases life expectancy
After a certain point, increasing GDP, life_expec grows very slowly.
3. Outliers:<br>
Countries with low GDP and low life expectancy are likely to be war-torn, have very poor healthcare, and have poor infrastructure

---

â˜‘ï¸ Conclusion:<br>
As GDP increases, life_expec increases, but after a certain point, it becomes dependent on other factors.

```python
sns.set_style(style="whitegrid")
plt.figure(figsize=(20, 10), dpi=120)
sns.scatterplot(data=df, x="income", y="child_mort", s=300)
plt.title(
    label="The relationship between countries' per capita income and mortality rates",
    fontsize=20,
)
plt.xlabel(xlabel="income", fontsize=18)
plt.ylabel(ylabel="child_mort", fontsize=18)
plt.tick_params(axis="both", labelsize=17)
plt.show()
```

âœ… General pattern:
1. As per capita income increases, the mortality rate decreases
2. The mortality rate stabilizes after income of $20,000
3. This means that beyond a certain annual income rate, other factors affect the mortality rate, but the annual income rate is also an important characteristic
4. Outliers:
    - We see 3 points:
        - Income above $120,000, these countries probably have high income, high exports, and a strong and stable economy.
        - With an income of between $20,000 and $40,000 and a mortality rate of 100 to 150, these countries have relatively good per capita incomes, but perhaps they have high mortality rates or poor health due to lack of investment in health.
        - A death rate above 200, this country probably has a weak economy, very poor healthcare.

In the next chart, we show the relationship between GDP per capita and mortality rate with the effect of the health feature with plotly.

```python
fig = px.scatter(
    df,
    x="gdpp",
    y="child_mort",
    color="health",
    color_continuous_scale="viridis",
    hover_name="country",
    size_max=40,
    opacity=0.7
)

fig.update_traces(marker=dict(size=22)) 
fig.update_traces(
    marker=dict(
        opacity=0.6,
        line=dict(width=1)
    )
)

fig.update_layout(
    title="Economic Development and Child Mortality: Evidence from Global Data",
    width=1400,
    height=700,
    title_font_size=22,
    xaxis_title="gdpp",
    yaxis_title="child_mort",
    font=dict(size=17),
    title_x = 0.5
    
)

fig.show()
```

The plot shows a strong negative and nonlinear relationship between GDP per capita and child mortality. Countries with higher income levels tend to have significantly lower child mortality rates. Additionally, the color gradient indicates that higher health expenditure is associated with better child survival outcomes. Even among countries with similar income levels, those with higher health spending generally exhibit lower mortality rates, suggesting that healthcare investment plays a critical role alongside economic development.
