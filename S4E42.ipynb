Countries with low GDP and low life expectancy are likely to be war-torn, have very poor healthcare, and have poor infrastructure

---

â˜‘ï¸ Conclusion:<br>

# model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
```

```python
x = df.drop(columns="country")
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()),
    ('kmeans', KMeans(random_state=42))
])
```

```python
from sklearn.metrics import silhouette_score
def silhouette_scorer(estimator, x):
    labels = estimator.fit_predict(x)
    return silhouette_score(x, labels)
```

```python
param_grid = {
    "pca__n_components": range(2, 8),
    "kmeans__n_clusters": range(2, 5)
}

grid = GridSearchCV(pipeline, param_grid, cv=5, scoring=silhouette_scorer)
grid.fit(x)

print(grid.best_params_)
```

```python
pipeline.set_params(pca__n_components = 2, kmeans__n_clusters = 2)
```

```python
pipeline.fit(X=x)
```

```python
labels = pipeline["kmeans"].labels_
```

```python
pipeline["kmeans"].inertia_
```

```python
silhouette_score(x, labels=labels)
```

```python
pca_data = pipeline["pca"].transform(
    pipeline["scaler"].transform(x)
)

pca_df = pd.DataFrame(pca_data, columns=["PC1", "PC2"])
pca_df["cluster"] = labels
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="cluster",
    palette="Set2",
    s=80
)
plt.title("Clusters in PCA Space")
plt.show()
```

```python
cluster_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("kmeans", KMeans(n_clusters=2, random_state=42))
])

cluster_pipe.fit(x)
labels = cluster_pipe["kmeans"].labels_
```

```python
centers_scaled = cluster_pipe["kmeans"].cluster_centers_
centers_real = cluster_pipe["scaler"].inverse_transform(centers_scaled)
centers_df = pd.DataFrame(centers_real, columns=x.columns)
```

```python
plt.figure(figsize=(8,6))


plt.scatter(
    df["income"],
    df["life_expec"],
    c=labels,
    cmap="viridis",
    alpha=0.5
)


plt.scatter(
    centers_df["income"],
    centers_df["life_expec"],
    s=300,
    c="red",
    marker="X",
    label="Centers"
)

plt.xlabel("income")
plt.ylabel("life_expec")
plt.title("Clusters with Centers")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.heatmap(
    centers_df,
    annot=True,
    fmt=".0f",
    cmap="coolwarm"
)
plt.title("Cluster Centers Heatmap", fontsize=17)
plt.xlabel("Features", fontsize=15)
plt.ylabel("Clusters", fontsize=15)
plt.tick_params(axis="both", labelsize=14)
plt.show()
```
